```python
# import everything from sympy moduleb 
from sympy import *

# pretty math formatting
init_printing()   # latex
```

```python
x, y, z = symbols('x y z')
```

```python
f = 2*x ** 3
f
```

```python
g = x**2 + x*y**2
g
```

## function evaluation

```python
f.subs(x, 0)
```

```python
f.subs(x, 1)
```

```python
g.subs([(x,2), (y,3)])
```

```python
# import everything from sympy moduleb 
from sympy import *

# pretty math formatting
init_printing()   # latex
```

```python
## multiple graph in same window

f = x**2
g = x**3

p = plot(f,g, (x, -2, 2),show=False, legend=True  )

p[0].line_color = 'blue'
p[1].line_color = 'green'


p[0].label      = 'f(x)'
p[1].label      = 'g(x)'
p.show()
```

```python
## undocking plots out of browser
%matplotlib qt  

f = x**2
plot(f)
```
