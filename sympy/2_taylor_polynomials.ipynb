```python
from sympy import *
init_printing()

%matplotlib inline
```

```python
x = symbols('x')
```

## Example1
find 4th degree Taylor polynomial of f(x) = sin(x) at a = pi/2.
plot f(x) and taylor polynomial on the same window

```python
## Finding Taylor polynomail

a = pi / 2
f = sin(x)
T = f.subs(x,a) \
    + diff(f,x,1).subs(x,a)*(x-a)  \
    + diff(f,x,2).subs(x,a)*(x-a)**2/2 \
    + diff(f,x,3).subs(x,a)*(x-a)**3/factorial(3) \
    + diff(f,x,4).subs(x,a)*(x-a)**4/factorial(4)
  
T
```

```python
## plotting f(x) and T(x) in same window


p = plot(f,T, (x,a - 4, a + 4),show=False, legend=True)

p[0].line_color = 'blue'
p[1].line_color = 'green'

p[0].label      = 'f(x)'
p[1].label      = 'T(x)'

p.show()
```

## Example 2

```python
## evaluation of T

a = 1
n = 20
f = log(x)

T = f.subs(x,a)

for k in range(1,n+1):
    df = diff(f,x,k)
    T = T + df.subs(x,a)*(x-a)**k/factorial(k)
    
T
```

```python
## plotting f(x) and T(x) in same window


p = plot(f,T, (x, 0.5, 2),show=False, legend=False)

p[0].line_color = 'blue'
p[1].line_color = 'green'

p[0].label      = 'f(x)'
p[1].label      = 'T(x)'

p.show()
```
